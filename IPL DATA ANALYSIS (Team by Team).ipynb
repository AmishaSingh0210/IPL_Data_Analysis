### Adding a column for Season

```python
data_frame_2020['Season'] = pd.DatetimeIndex(data_frame_2020['date']).year
data_frame_2020.head()
```

### Merging Data

```python
season_data=data_frame_2020[['id','Season']].merge(data_balls2020, left_on = 'id', right_on = 'id', how = 'left').drop('id', axis = 1)
season_data.head()
```

### Performance of Batsman in each season 

```python
Batsman = season_data[season_data['batsman']=='AM Rahane']
SRuns_batsman= Batsman.groupby('Season')['batsman_runs'].sum().reset_index()
p=SRuns_batsman.set_index('Season')
plt.figure(figsize=(15,8))
ax = plt.axes()
ax.set(facecolor = "white")
sns.lineplot(data=p,palette="magma") 
plt.title('Runs by Ajinkya Rahane',fontsize=12,fontweight="bold")
plt.show()
```

### Performance of Bowler 

```python
Bowler = season_data[season_data['bowler']=='SP Narine']
wickets= Bowler.groupby('Season')['is_wicket'].sum().reset_index()
p=wickets.set_index('Season')
plt.figure(figsize=(15,8))
ax = plt.axes()
ax.set(facecolor = "white")
sns.lineplot(data=p,palette="magma") 
plt.title('Wickets by Sunil Narine',fontsize=12,fontweight="bold")
plt.show()
```

### Dismissal Kind for a player 

```python
player = (data_balls2020['batsman']=='MS Dhoni')
df_dhoni=data_balls2020[player]
plt.figure(figsize=(10,10))
df_dhoni['dismissal_kind'].value_counts().plot.pie(autopct='%1.1f%%',rotatelabels=True)
plt.title("Dismissal Kind",fontweight="bold",fontsize=15)
plt.show()
```

### Runs scored by a player 

```python
def count(df_dhoni,runs):
    return len(df_dhoni[df_dhoni['batsman_runs']==runs])*runs
print("Runs scored from 1's :",count(df_dhoni,1))
print("Runs scored from 2's :",count(df_dhoni,2))
print("Runs scored from 3's :",count(df_dhoni,3))
print("Runs scored from 4's :",count(df_dhoni,4))
print("Runs scored from 6's :",count(df_dhoni,6))
```

### Highest Strike Rate in Death Overs 

```python
death_overs = data_balls2020[data_balls2020['over']>15]
death_overs.groupby('batsman')['batsman_runs'].count()>200
all_batsman = death_overs.groupby('batsman')['batsman_runs'].count()
x = all_batsman>200
batsman_list = all_batsman[x].index.tolist()
final = data_balls2020[data_balls2020['batsman'].isin(batsman_list)]
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
strike_rate = runs*100/balls
strike_rate.sort_values(ascending= False)
```
