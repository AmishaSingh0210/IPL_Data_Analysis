### Adding a column for Season

```python
data_frame_2020['Season'] = pd.DatetimeIndex(data_frame_2020['date']).year
data_frame_2020.head()
```

### Merging Data

```python
season_data=data_frame_2020[['id','Season']].merge(data_balls2020, left_on = 'id', right_on = 'id', how = 'left').drop('id', axis = 1)
season_data.head()
```

### Performance of Batsman in each season 

```python
VK = season_data[season_data['batsman']=='V Kohli']
SRuns_batsman= VK.groupby('Season')['batsman_runs'].sum().reset_index()
p=SRuns_batsman.set_index('Season')
plt.figure(figsize=(15,8))
ax = plt.axes()
ax.set(facecolor = "white")
sns.lineplot(data=p,palette="magma") 
plt.title('Runs by Virat Kohli',fontsize=12,fontweight="bold")
plt.show()
```

### Toss Decision Across Different cities 

```python
plt.figure(figsize=(15,8))
ax = plt.axes()
ax.set(facecolor = "white")
sns.countplot(x='city', hue='toss_decision', data=data_frame_2020,palette="magma",saturation=1)
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=15)
plt.xlabel('\n City',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title('Toss decision across different cities',fontsize=12,fontweight="bold")
plt.show()
```

### Runs scored by a player 

```python
def count(df_dhoni,runs):
    return len(df_dhoni[df_dhoni['batsman_runs']==runs])*runs
print("Runs scored from 1's :",count(df_dhoni,1))
print("Runs scored from 2's :",count(df_dhoni,2))
print("Runs scored from 3's :",count(df_dhoni,3))
print("Runs scored from 4's :",count(df_dhoni,4))
print("Runs scored from 6's :",count(df_dhoni,6))
```

### Top 10 run scorer in IPL

```python
runs = data_balls2020.groupby(['batsman'])['batsman_runs'].sum().reset_index()
runs.columns = ['Batsman', 'runs']
y = runs.sort_values(by='runs', ascending = False).head(10).reset_index().drop('index', axis=1)
ax = plt.axes()
ax.set(facecolor = "white")
sns.barplot(x=y['Batsman'],y=y['runs'],palette='rocket',saturation=1)
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('\n Player',fontsize=15)
plt.ylabel('Total Runs',fontsize=15)
plt.title('Top 10 run scorers in IPL',fontsize=15,fontweight="bold")
```

### Which player has hit most number of fours

```python
fours = data_balls2020['batsman_runs']==4
fours_count = data_balls2020[fours]
fours_count.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
```

### Highest Strike Rate in Death Overs 

```python
death_overs = data_balls2020[data_balls2020['over']>15]
death_overs.groupby('batsman')['batsman_runs'].count()>200
all_batsman = death_overs.groupby('batsman')['batsman_runs'].count()
x = all_batsman>200
batsman_list = all_batsman[x].index.tolist()
final = data_balls2020[data_balls2020['batsman'].isin(batsman_list)]
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
strike_rate = runs*100/balls
strike_rate.sort_values(ascending= False)
```

### Which team has hit how many number of sixes in each over
